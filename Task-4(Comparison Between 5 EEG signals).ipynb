```python
import pandas as pd
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = pd.read_csv('ica.csv')
df = pd.DataFrame(file)
df.drop(df.columns[0],inplace=True,axis=1)
df.head(3)
```

```python
for i in range(len(df.columns)):
    plt.figure(figsize=(17,5))
    plt.subplot(1,2,1)
    plt.specgram(df[df.columns[i]][:499201],Fs=(1/128),sides='onesided',mode='magnitude',scale='dB',cmap=plt.get_cmap('vlag'))
    plt.xlabel('Time')
    plt.ylabel('Freq')
    plt.colorbar()
    plt.title('Spectrogrum of The channel named(class-2) : {}'.format(df.columns[i]))
    
    plt.subplot(1,2,2)
    plt.specgram(df[df.columns[i]][499201:],Fs=(1/128),sides='onesided',mode='magnitude',scale='dB',cmap=plt.get_cmap('vlag'))
    plt.xlabel('Time')
    plt.ylabel('Freq')
    plt.colorbar()
    plt.title('Spectrogrum of The channel named(class-1) : {}'.format(df.columns[i]))
    
    plt.show()
```
