```python
import pandas as pd
import numpy as np
from sklearn.decomposition import FastICA
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = pd.read_csv('Raw_TrainTest_MultiTasks.csv')
df = pd.DataFrame(file)
df.head()
```

```python
ica = FastICA()
```

```python
df_ = pd.DataFrame(ica.fit_transform(df.drop('class_',axis=1)),columns=df.columns[:-1])
df_ = pd.concat([df_,df['class_']],axis=1)
df_.head()
```

```python
df_2 = pd.DataFrame(df_[df_['class_']==2])
df_2.head()
```

```python
df_1 = pd.DataFrame(df_[df_['class_']==1])
df_1.head()
```

```python
print("DataFrame with class 2 :",df_2.shape," AND  ","DataFrame with class 1 :",df_1.shape)
```

```python
for i in range(len(df_.columns)-1):
    plt.figure(figsize=(17,5))

    plt.subplot(1,3,1)
    x = np.arange(0,422400/128,(1/128))
    plt.plot(x,df_1[df_1.columns[i]],'blue',label='before SMA')
    x = np.arange(0,422400/128,(1/128))
    file = pd.DataFrame(df_1[df_1.columns[i]].rolling(500).mean())
    plt.plot(x,file,'red',label='after SMA')
    plt.legend(loc='best')
    plt.title('CLASS 1 of '+df_.columns[i])
    
    
    plt.subplot(1,3,2)
    x = np.arange(0,499200/128,(1/128))
    plt.plot(x,df_2[df_2.columns[i]],'blue',label='before SMA')
    file = pd.DataFrame(df_2[df_2.columns[i]].rolling(500).mean())
    x = np.arange(0,499200/128,(1/128))
    plt.plot(x,file,'red',label='after SMA')
    plt.legend(loc='best')
    plt.title('CLASS 2 of '+df_.columns[i])
    
    
    plt.subplot(1,3,3)
    x = np.arange(0,7200,(1/128))
    plt.plot(x,df_[df_.columns[i]],'blue',label='before SMA')
    file = pd.DataFrame(df_[df_.columns[i]].rolling(500).mean())
    x = np.arange(0,7200,(1/128))
    plt.plot(x,file,'red',label='after SMA')
    plt.legend(loc='best')
    plt.title('BOTH CLASS')
```

```python
df_1_new = pd.DataFrame()
df_2_new = pd.DataFrame()
```

```python
li = ['F3','O1','F8']

for i in range(len(li)):
    file = pd.DataFrame(df_1[li[i]].rolling(5000).mean(),columns=[li[i]])
    df_1_new = pd.concat([df_1_new,file],axis=1)
    
    file = pd.DataFrame(df_2[li[i]].rolling(5000).mean(),columns=[li[i]])
    df_2_new = pd.concat([df_2_new,file],axis=1)
```

```python
df_2_new.dropna(axis=0,inplace=True)
df_1_new.dropna(axis=0,inplace=True)
```

```python
df_1_new = df_1_new.reset_index(drop=True)
df_2_new = df_2_new.reset_index(drop=True)
```

```python
file = pd.DataFrame(np.array([1]*df_1_new.shape[0]),columns=['class_'])
df_1_new = pd.concat([df_1_new,file],axis=1)
```

```python
file = pd.DataFrame(np.array([2]*df_2_new.shape[0]),columns=['class_'])
df_2_new = pd.concat([df_2_new,file],axis=1)
```

```python
df = pd.concat([df_1_new,df_2_new],axis=0)
```

```python
df.to_csv('avg_.csv')
```
