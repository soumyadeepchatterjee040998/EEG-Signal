```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.fftpack import fft,fftshift,fftfreq,ifft
from scipy import signal
from scipy import stats
```

```python
file = pd.read_csv('gamma.csv')
df = pd.DataFrame(file)
df.drop(df.columns[0],axis=1,inplace=True)
df.head(3)
```

```python
x = np.arange(0,7200,(1/128))

for i in range(len(df.columns)):
    
    plt.figure(figsize=(20,5))
    plt.subplot(1,2,1)
    
    plt.plot(x[:499201],df[df.columns[i]][:499201],'go')    
    plt.xlabel('T(time)')
    plt.ylabel('Amp')
    plt.title('Signal in Time Domain(CLASS-2)')
    
    plt.subplot(1,2,2)
    plt.plot(x[499201:],df[df.columns[i]][499201:],'yo')
    plt.xlabel('T(time)')
    plt.ylabel('Amp')
    plt.title('Signal in Time Domain(CLASS-1)')
    plt.show()

    print("This is the GAMMA Band of ",df.columns[i])
    print('summery :(class-2) ',' (class-1)')

    print(pd.concat([pd.DataFrame(df[df.columns[i]][:499201].describe()),pd.DataFrame(df[df.columns[i]][499201:].describe())],axis=1))
```

```python
b,a = signal.butter(10,[38*(2*np.pi),42*(2*np.pi)],btype='bandpass',analog=True,output='ba')
```

```python
for i in range(len(df.columns)):
    plt.figure(figsize=(17,5))
    plt.subplot(1,3,1)
    
    w2,h2 = signal.freqs(b,a,worN=499201)
    y2 = np.abs(fft(np.array(df[df.columns[i]][:499201]))*h2)
    plt.plot(w2,y2,'y',label='class-2')
    plt.xlabel("W(rad/s)")
    plt.ylabel("Amp")
    plt.title('Amplitude spectrum(class-2)')
    
    plt.subplot(1,3,2)
    w1,h1 = signal.freqs(b,a,worN=(921600-499201))
    y1 = np.abs(fft(np.array(df[df.columns[i]][499201:]))*h1)
    plt.plot(w1,y1,'b',label='class-1')
    plt.xlabel("W(rad/s)")
    plt.title('Amplitude spectrum(class-1)')
  
    plt.subplot(1,3,3)
    plt.plot(w2,y2,'y',label='class-2')
    plt.plot(w1,y1,'b',label='class-1')
    plt.xlabel("W(rad/s)")
    plt.title('Amplitude spectrum')
    plt.legend(loc='best')
    
    plt.show()
    print("This is the GAMMA Band of ",df.columns[i])
```

```python
for i in range(len(df.columns)):
    plt.figure(figsize=(17,5))
    
    plt.subplot(1,3,1)
    w2,h2 = signal.freqs(b,a,worN=499201)
    y2 = 20*np.log10(np.abs(fft(np.array(df[df.columns[i]][:499201]))*h2)**2)
    plt.plot(w2,y2,'y',label='class-2')
    plt.xlabel("W(rad/s)")
    plt.ylabel("Amp(in db(20*log10))")
    plt.title('PSD(class-2)')
    
    plt.subplot(1,3,2)
    w1,h1 = signal.freqs(b,a,worN=(921600-499201))
    y1 = 20*np.log10(np.abs(fft(np.array(df[df.columns[i]][499201:]))*h1)**2)
    plt.plot(w1,y1,'b',label='class-1')
    plt.xlabel("W(rad/s)")
    plt.title('PSD(class-1)')
  
    plt.subplot(1,3,3)
    plt.plot(w2,y2,'y',label='class-2')
    plt.plot(w1,y1,'b',label='class-1')
    plt.xlabel("W(rad/s)")
    plt.title('PSD')
    plt.legend(loc='best')
    
    plt.show()
    print('This is the PSD of GAMMA Band of ',df.columns[i])
```

```python
for i in range(len(df.columns)):
    plt.figure(figsize=(20,5))
    plt.subplot(1,2,1)
    
    w2,h2 = signal.freqs(b,a,worN=499201)
    y2 = np.angle(fft(np.array(df[df.columns[i]][:499201]))*h2)
    plt.plot(w2,y2,'y')
    plt.xlabel('W(rad/sec)')
    plt.ylabel('Phase(rad)')
    plt.title('Phase spectrum(class-2)')
    
    plt.subplot(1,2,2)
    w1,h1 = signal.freqs(b,a,worN=(921600-499201))
    y1 = np.angle(fft(np.array(df[df.columns[i]][499201:]))*h1)
    plt.plot(w1,y1,'b')
    plt.xlabel("W(rad/s)")
    plt.ylabel('Phase(rad)')
    plt.title('Phase spectrum(class-1)')
    
    plt.show()
    print("This is the GAMMA Band of ",df.columns[i])
```
